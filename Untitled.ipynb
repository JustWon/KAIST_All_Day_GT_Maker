```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
am02 = pd.read_csv('d:\Datasets\KAIST_All_Day\AM02\East\DGPS-IMU_V001.dat', sep=' ',header=None)
am05 = pd.read_csv('d:\Datasets\KAIST_All_Day\AM05\East\DGPS-IMU_V001.dat', sep=' ',header=None)
am09 = pd.read_csv('d:\Datasets\KAIST_All_Day\AM09\East\DGPS-IMU_V001.dat', sep=' ',header=None)
```

```python
standard = 1000
search_range = 50

dist_list = []
for i in range(standard-search_range,standard+search_range):
    dist_list.append(np.linalg.norm(am05.loc[standard] - am09.loc[i]))
    
plt.plot(dist_list)
plt.show()
```
